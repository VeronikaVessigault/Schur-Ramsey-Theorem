```python
# DO NOT EDIT
def get_list_csv(filename):
    file_list = ["color_classes"]
    if os.path.isfile("{}.csv".format(filename)) and filename in file_list:
        with open('{}.csv'.format(filename)) as csv_file:
            reader = csv.reader(csv_file)
            mylist = list(reader)
            list_tuples = []
            for i in mylist[0]:
                list_tuples.append(eval(i))
            return list_tuples
    return list()


def set_list_csv(filename, mylist):
    file_list = ["color_classes"]
    if os.path.isfile("{}.csv".format(filename)) and filename in file_list:
        with open('{}.csv'.format(filename), 'w') as csv_file:  
            writer = csv.writer(csv_file)
            writer.writerow(mylist)
            print("file written")
```

```python
# DO NOT EDIT
def get_dict_csv(filename):
    file_list = ["anti_schur_avg", "anti_schur_max", "anti_schur_runs", "schur_avg", "schur_max", "schur_runs"]
    if os.path.isfile("{}.csv".format(filename)) and filename in file_list:
        with open('{}.csv'.format(filename)) as csv_file:
            reader = csv.reader(csv_file)
            mydict = dict(reader)
            return mydict

def set_dict_csv(filename, mydict):
    file_list = ["anti_schur_avg", "anti_schur_max", "anti_schur_runs", "schur_avg", "schur_max", "schur_runs"]
    if os.path.isfile("{}.csv".format(filename)) and filename in file_list:
        with open('{}.csv'.format(filename), 'w') as csv_file:  
            writer = csv.writer(csv_file)
            for key, value in mydict.items():
                writer.writerow([key, value])
```

```python
# DO NOT EDIT
def reset_stats(all_stats = True, specific = -1):
    file_list = ["anti_schur_avg", "anti_schur_max", "anti_schur_runs", "schur_avg", "schur_max", "schur_runs"]
    keys = get_list_csv("color_classes")
    my_dict = {i:0 for i in keys}
    if all_stats:
        for i in file_list:
            set_dict_csv(i, my_dict)
        print("all stats reset")
    elif specific in file_list:
        set_dict_csv(specific, my_dict)
        print("stats for " + specific + " reset")
```

```python
def write_stats(runs = -1, r = -1, avg = -1, maximum = -1, anti = False):
    keys = get_list_csv("color_classes")
    #print("keys " + str(keys) + " type: " + str(type(keys[0])))
    #print("r: " + str(r) + " type: " + str(type(r)))
    if r not in keys:
        return False
    if not anti:
        dict_runs = get_dict_csv("schur_runs")
        dict_avg = get_dict_csv("schur_avg")
        dict_max = get_dict_csv("schur_max")
        #print(dict_runs)
        #print(dict_avg)
        #print(dict_max)
        
        # add new runs 
        old_runs = int(dict_runs[str(r)])
        new_runs = old_runs + runs
        dict_runs[str(r)] = new_runs
        
        # adjust new max
        if maximum > int(dict_max[str(r)]):
            dict_max[str(r)] = maximum
            
        # adjust average max
        old_avg = float(dict_avg[str(r)])
        old_avg_times_old_runs = old_avg * old_runs
        new_avg_times_runs = avg * runs
        adjusted_avg = (old_avg_times_old_runs+new_avg_times_runs)/new_runs
        dict_avg[str(r)] = adjusted_avg
        
        set_dict_csv("schur_runs", dict_runs)
        set_dict_csv("schur_avg", dict_avg)
        set_dict_csv("schur_max", dict_max)
        
        return True
        
    elif anti:
        dict_runs = get_dict_csv("anti_schur_runs")
        dict_avg = get_dict_csv("anti_schur_avg")
        dict_max = get_dict_csv("anti_schur_max")
        #print(dict_runs)
        #print(dict_avg)
        #print(dict_max)
        
        # add new runs 
        old_runs = int(dict_runs[str(r)])
        new_runs = old_runs + runs
        dict_runs[str(r)] = new_runs
        
        # adjust new max
        if maximum > int(dict_max[str(r)]):
            dict_max[str(r)] = maximum
        
        # adjust average max
        old_avg = float(dict_avg[str(r)])
        old_avg_times_old_runs = old_avg * old_runs
        new_avg_times_runs = avg * runs
        adjusted_avg = (old_avg_times_old_runs+new_avg_times_runs)/new_runs
        dict_avg[str(r)] = adjusted_avg
        
        set_dict_csv("anti_schur_runs", dict_runs)
        set_dict_csv("anti_schur_avg", dict_avg)
        set_dict_csv("anti_schur_max", dict_max)
        
        return True
```

```python
# DO NOT EDIT
# Given a r partitions, this checks if there is a "monochromatic sum" in any one of them.
def check_for_sum(partitions, r = -1, silence = False):
    for i in range(r):
        length = len(partitions[i])
        for j in range(length):
            for k in range(j, length):
                if partitions[i][j] + partitions[i][k] in partitions[i]:
                    if not silence:
                        print("Partition #" + str(i) + " got the sum  of " + str(partitions[i][j]) + " + " + str(partitions[i][k]), end=" ")
                    return i, partitions[i][j], partitions[i][k], True
    return -1, -1, -1, False
```

```python
# DO NOT EDIT
# Creates and populates a r partitions until a sum appears. Returns the integer which created the sum.
def continue_until_sum(r = -1, doubles = True, silence = False):
    sum = False
    n = 0
    partitions=[[] for i in range(r)]
    while not sum:
        n += 1
        rand_num = cryto_gen_num.randrange(r)
        partitions[rand_num].append(n)
        part_with_sum, x, y, sum = check_for_sum(partitions, r, silence = silence)
    if not silence:
        print("at integer " + str(n) + ". Here are the partitions: " + str(partitions) + ".")
    return n
```

```python
# DO NOT EDIT
# Runs the command above a set number of times. Collects all the integers which created the sum to return the mean and maximum of these integers.
def run_many_continue_until_sum(runs = -1, r = -1, doubles = True):
    ending_nums = []
    for i in range(runs):
        n = continue_until_sum(r = r, silence = True)
        ending_nums.append(n)
    if runs <= 500:
        print("List: " + str(ending_nums))
    mean = fmean(ending_nums)
    print("Average: " + str(mean))
    maximum = max(ending_nums)
    print("Maximum: " + str(maximum))
    success = write_stats(runs = runs, r = r, avg = mean, maximum = maximum, anti = False)
    if success:
        success = True
        #print("stats successfully written")
    else:
        print("stats did not write successfully")
    return mean, maximum
```

```python
def anti_sum_schur(r = -1, doubles = True):
    forced_sum = False
    n = 0
    partitions=[[] for i in range(r)]
    while not forced_sum:
        n += 1
        good_parts = valid_non_sum_choices(partitions, n, r)
        if good_parts:
            rand_num = cryto_gen_num.randrange(len(good_parts))
            partitions[good_parts[rand_num]].append(n)
        else:
            forced_sum = True
    #print(n, partitions)
    return n
```

```python
def valid_non_sum_choices(partitions, n, r = -1):
    good_parts = []
    for i in range(r):
        good = check_sum_in_partition(partitions, n, i)
        if good:
            good_parts.append(i)
    #print(good_parts)
    return good_parts
```

```python
# DO NOT EDIT
def check_sum_in_partition(partitions, n, part = -1):
    length = len(partitions[part])
    for i in range(length):
        for j in range(i, length):
            if (partitions[part][i] + partitions[part][j]) == n:
                #print(part, partitions[part][i], partitions[part][j], True)
                return False
    return True
```

```python
# DO NOT EDIT
# Runs the command above a set number of times. Collects all the integers which forced the sum so that the mean and maximum of these integers can be returned.
def run_many_until_sum_anti(runs = -1, r = -1, doubles = True):
    ending_nums = []
    for i in range(runs):
        ending = anti_sum_schur(r = r, doubles = True)
        ending_nums.append(ending)
        if runs >= 20000:
            if i == runs//2:
                print("50% done!")
            elif runs >= 50000:
                if i == runs//4:
                    print("25% done!")
                elif i == runs//4*3:
                    print("75% done!")
    if runs <= 500:
        print("List: " + str(ending_nums))
    mean = fmean(ending_nums)
    print("Average: " + str(mean))
    maximum = max(ending_nums)
    print("Maximum: " + str(maximum))
    success = write_stats(runs = runs, r = r, avg = mean, maximum = maximum, anti = True)
    if success:
        success = True
        #print("stats successfully written")
    else:
        print("stats did not write successfully")
    return mean, maximum
```
